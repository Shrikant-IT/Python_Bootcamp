### List Comprehension  

```python
## There are two type of comprehension in python list and dictonary.
```

```python
["Rashu" for i in range(0, 5)]
```

```python
["$" for i in range (1, 3)]
```

```python
## condition within list comprehension
## syntax -->  [<expression> <loop> <condition>]
```

```python
# return list of number from 1-50 which are less than 5

[i for i in range(1, 51) if i <= 4 ]
```

```python
# return squares of only odd numbers from 1-10 numbers

[x*x for x in range(1, 11) if x %2 != 0]
```

```python
## if else condition within list compreshension 
## syntax ---> [<output from if > <if condition> <else> <else result> <loop>]
```

```python
marks = [23,56,45,12]

# create list with pass / fail based on 35 marks passing criteria
```

```python
["Pass" if m>= 35 else "Fail" for m in marks]
```

```python
data = [-6,0,3,5,-2,0]
```

```python
["Zero" if d==0  else "Negative" if d<0 else "Positive" for d in data]
```

```python
fruits = ['apple', 'cherry', 'orange']
colors = ['green','red','yellow']
```

```python
for a in fruits:
    for b in colors:
        print(a, b)
```

```python
### zip function in python--> its return one one elements from the each list
```

```python
for a, b in zip(fruits, colors):
    print(b,a)
```
