```python
import traceback

a = 10
b = 2
n = [2, 5, 25, 35, 11]

try:
    c=a/b
    n[2]
except Exception:
    print(traceback.format_exc())
    print(n)
else:
    print("Will be executed if not exeecption is thrown")
finally:
    print("Always be executed...if or not an exception is thrown")
print(c)
print(n)
```

```python
import traceback
a=10
b=0
n=[2,3,123,211]
try:
    c=a/b
    n[15]
except ZeroDivisionError:
    print(traceback.format_exc())
except IndexError:
    print(traceback.format_exc())
else:
    print("Will be executed if not exeecption is thrown")
finally:
    print("Always be executed...if or not an exception is thrown")
print("byee")
print("done")
```
