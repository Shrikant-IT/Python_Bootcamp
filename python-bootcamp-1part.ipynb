```python
# Rules for defining the variable

1. Variable can not be started with any number
    12a2 (Not Correct)
2. Numbers are allowed in between and in end
    num1 (Correct)
    num1num (Correct)
3. None of the symbols are allowed in variable name other than underscore'_'
    w@3 (Not Correct)
    num_1 (Correct)
    num 1 (Not Correct)
4. Capital case and small case has different meaning.
    x is not equal to X
```
