## Group By 
## important fuctiion 

## It is used to see the aggregated results on the dataframe

```python
for i in range(1,13):
    print(i, round(df[df["Salary"]==i]["Team"].mean(),2))
```

```python
df.groupby("Salary")["Bonus %"].mean()
#mnth : Grouping Column
#temp : Aggregating Column
#mean : Aggregating Function
```

```python
df.groupby("Team")[["Salary"]].mean()
```

```python
df.groupby("Bonus %")[["Salary"]].median()
```

```python
d={"Salary":["max","min","mean"],"Bonus %":["median","std"]}
df.groupby("Salary").agg(d)
```
