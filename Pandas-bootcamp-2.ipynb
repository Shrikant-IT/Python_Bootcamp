```python
import pandas as pd
import numpy as np
```

```python
# read the files

df = pd.read_csv("telecom_churn.csv")
```

```python
df.head()
```

```python
# filtering data bsed on conditions

# df[df[column name]< condition operator> value]
```

```python
df['Churn'].value_counts()
```

```python
df[df['Churn']==True]
```

```python
# filtering data based on multiple columns

df[(df['Churn']==True) & (df['Area code']==408)]
```

```python
# cross tab : it gives calculations between two or more rows

pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], [df['International plan'], df['Area code'], df['Voice mail plan']])
```

```python
pd.crosstab(df['Churn'], df['State'])
```

```python
# groupby  : 

# df.groupby(grouping column)[columns to show].function()
```

```python
column_to_show = ['Total day calls','Total night calls']

df.groupby('Churn')[column_to_show].mean()
```

```python
# internal process

df[(df['Churn']==False)].iloc[:, ['Total day calls', 'Total night calls']].mean()
```

```python
# another method for groupby :.agg


df.groupby('Area code').agg({'Total night calls': 'mean'})
```

```python
df.groupby('Area code').agg({'Total day calls': 'min', 'Total night calls': 'mean'})
```

```python
df.groupby('Area code').agg({'Total day calls': np.min, 'Total night calls': np.mean})
```

```python
# groupby on multiple columns
```

```python
df.groupby(['Churn','Area code']).agg({'Total day calls':np.max,'Total night calls':np.mean})
```

```python
# merge
# create random data.
```

```python
std_details = {'id':[1,2,3,4,5],
               'name':['A','B','C','D','E'],
              'region':['N','S','W','W','S']}
stu_marks = {'id':[1,2,3,4,6],
               'sub':['py','matlab','C','R','java'],
              'marks':[10,13,14,16,13]}


df1 = pd.DataFrame(std_details)
df2 = pd.DataFrame(stu_marks)
```

```python
df1
```

```python
df2
```

```python
# inner merge : default

pd.merge(df1, df2)
```

```python
# standard method

pd.merge(df1, df2, on = 'id')
```

```python
# left merge : fetch everything from left df but common one from right df

pd.merge(df1, df2, on = 'id', how='left')
```

```python
# right merge : fetch everything from right df but common one from left df

pd.merge(df1, df2, on = 'id', how='right')
```

```python
# outer : union of both the tables

pd.merge(df1, df2, on = 'id', how='outer')
```

```python
# when reference column names are diff in two tables

std_details1 = {'s_id':[1,2,3,4,5],
               'name':['A','B','C','D','E'],
              'region':['N','S','W','W','S']}
stu_marks1 = {'id':[1,2,3,4,6],
               'sub':['py','matlab','C','R','py'],
              'marks':[10,13,14,16,13]}


s1 = pd.DataFrame(std_details1)
s2 = pd.DataFrame(stu_marks1)
```

```python
s1
```

```python
s2
```

```python
# pd.merge(df3,df4) : it will throw an error
```

```python
pd.merge(s1, s2, left_on='s_id', right_on='id')
```

```python
pd.merge(s1, s2, left_on='s_id', right_on='id', how='left')
```

```python
pd.merge(s1, s2, left_on='s_id', right_on='id', how='right')
```

```python
pd.merge(s1, s2, left_on='s_id', right_on='id', how='outer')
```
